### Functions for processing

To read to list

```python
def read_csv_to_list(file_path, index_start_reading, colum_number):
    return_array = list()
    with open(file_path, newline='') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
        for index, row in enumerate(spamreader):
            if index >= index_start_reading:
                return_array.append(float(row[colum_number]))
    return return_array
```

Calculate drop/succ rate 

```python
def calculate_drop_succ_rate(drop_arr, succ_arr):
    rate_arr = [drop_arr[i]/(drop_arr[i] + succ_arr[i]) for i in range(len(drop_arr))]
    return rate_arr
```

```python
def calculate_drop_total_rate(drop_arr, total_arr):
    rate_arr = [drop_arr[i]/total_arr[i] for i in range(len(drop_arr))]
    return rate_arr
```

Filter for abnormal value

```python
def filter_abnormal_values(array, value):
    return [i for i in array if i < value]
```

Ingress normalize

```python
def ingress_traffic_calculation(ingress_arr):
    return [i*500 for i in ingress_arr]
```

Calculate mean

```python
def calculate_mean(arr):
    return sum(arr)/len(arr)
```

### Read files to variables


```python
drop_request_ing = dict()
succ_request_ing = dict()
drop_succ_rate_ing = dict()
reward_sim = list()

algos = ['trained_400', 'random', 'shortest', 'balance']
topology = "octagon"
ingress_node_list = ["2node", "4node", "6node", "8node"]
for ingress_node in ingress_node_list:
    drop_request = list()
    total_request = list()
    drop_succ_rate = list()
    reward_sim.append(read_csv_to_list(eval(f"{topology}_{ingress_node}_reward_file_path"), 2, 0))
    for ind, algo in enumerate(algos):
        drop_request.append(read_csv_to_list(eval(f"{algo}_{topology}_{ingress_node}_drop_request"), 1, 3))
        total_request.append(read_csv_to_list(eval(f"{algo}_{topology}_{ingress_node}_total_request"), 1, 4))
        drop_succ_rate.append(calculate_drop_total_rate(eval(f"drop_request[{ind}]"), eval(f"total_request[{ind}]")))
    drop_request_ing[ingress_node] = drop_request
    succ_request_ing[ingress_node] = total_request
    drop_succ_rate_ing[ingress_node] = drop_succ_rate
    
```

```python
avg_drop_rate_ing = dict()

for ind, algo in enumerate(algos):
    avg_drop_rate = list()
    for ingress_node in ingress_node_list:
        avg_drop_rate.append(calculate_mean(drop_succ_rate_ing[ingress_node][ind]))
    avg_drop_rate_ing[algo] = avg_drop_rate
```

```python
for ind, ingress_node in enumerate(ingress_node_list):
    reward_sim[ind].pop()
    plt.plot(reward_sim[ind], label=f'{ingress_node}')
plt.title("Reward of RL models in different client node number")
plt.xlabel("Trained episodes")
plt.ylabel("Reward")
plt.legend()
plt.show()
```
